```python
# Load selenium components
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys 

# Establish chrome driver and go to report site URL
url = "https://stage.travis.prodigycad.com/property-search"
#driver = webdriver.Chrome('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

#driver = webdriver.Chrome(executable_path='/usr/local/lib/python3.7/dist-packages/selenium/webdriver/common/service.py')
driver.get(url)
#is this up to date
#test to see if site accessed
try:
  element = WebDriverWait(driver, 10).until(
      EC.presence_of_element_located((By.ID, "searchInput"))
  )
  print("it's working!")
except TimeoutException:
  print("not loaded!")
#lookin for the property search field
searchInput = driver.find_element(By.ID, "searchInput")
#search for the value in send_keys(searchValue)
searchInput.send_keys('1808')
#submit search
driver.find_element(By.CSS_SELECTOR,'button.MuiButtonBase-root.MuiIconButton-root.jss4').click()
#test to scrape table data
try:
  element = WebDriverWait(driver, 10).until(
      EC.presence_of_element_located((By.CSS_SELECTOR,'div.ag-row.ag-row-no-focus.ag-row-odd.ag-row-level-0.ag-row-position-absolute'))
  )
  print("found the rows!")
except TimeoutException:
  print("no rows!")
rows = driver.find_elements(By.CSS_SELECTOR,'div.ag-row.ag-row-no-focus.ag-row-odd.ag-row-level-0.ag-row-position-absolute')
#years = driver.find_elements(By.CSS_SELECTOR,'div.ag-cell.ag-cell-not-inline-editing.ag-cell-auto-height.ag-cell-value')
owners = driver.find_elements(By.XPATH,"//div[@id='root']/div/div/div/div/div[2]/div/div[2]/div[3]/div[2]/div/div/div[2]/div[2]/div[3]/div[2]/div/div/div/div")
#css=.ag-cell-focus

#items = driver.find_elements(By.CSS_SELECTOR,'div.ag-row.ag-row-no-focus.ag-row-odd.ag-row-level-0.ag-row-position-absolute')
for year in owners:
  print(year.text)
driver.close()

#Use the selenium IDE to help you, seriously
#element = driver.find_element(By.TAG_NAME, "span")
#elements = driver.find_element(By., "//input[@id='searchInput']")

#print(element)
#for e in elements:
  #print(e.text)

```
